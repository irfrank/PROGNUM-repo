```python
import numpy as np
def stdev(arr):
    arr=np.array(arr)
    mean=np.sum(arr)/ len(arr)
    o=np.sum((arr-mean)**2)/len(arr)
    return np.sqrt(o)

x=np.array([1,2,3,4,5]) #example array

print (stdev(x))
print (np.std(x)) 

def stdev1(arr):
    arr=np.array(arr)
    return np.sqrt(np.mean((arr-np.mean(arr))**2))
print (stdev1(x))
```

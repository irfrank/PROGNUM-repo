```python
night1 = [1,3,2,4,0,3,7,5,2,9,2,4,3]
A = set(night1)
night2 = [0,0,2,4,2,4,3,2,1,1,1,2,0]
B = set(night2)
inters = set.intersection(A,B)
C = list(inters)
print(f"Intersection night1, night2: {C}")

if 5 in C:
    print("5 is in the intersection list")
else:
    print("5 is not in the intersection list")

if 5 in A:
    print ("5 is a member of the set version of night1")
```

```python
from array import *
import numpy as np

mu, sigma = 0.0, 1.4 # mean and standard deviation

s = np.random.normal(mu, sigma, 100000) #function
abs(mu - np.mean(s))
standard_deviation= abs(np.std(s, ddof=1)) 

print (f"the standard deviation is {standard_deviation}")
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *np.exp( - (bins - mu)**2 / (2 * sigma**2) ),linewidth=2, color='r')
plt.show()
```
